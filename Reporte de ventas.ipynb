```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
```

```python
#leer excel como df
df = pd.read_excel('Reporte de ventas por articulos MOD 3.xlsx')
```

```python
#corregir la columna Ganancia Prom. % para que los numeros no se vean asi 2.955130e+05 y tranformar los valores a tipo int
df['Ganancia Prom. %'] = df['Ganancia Prom. %'].apply(lambda x: f'{x:.2f}')
```

```python
#tranformar a str las columnas Nombre del Producto y Categoria
df['Nombre del Producto'] = df['Nombre del Producto'].astype(str)
df['Categoria'] = df['Categoria'].astype(str)
```

```python
# Filtrar y ordenar datos
df_sorted = df[df['Categoria'] == 'Auricular'].nlargest(10, 'Cantidad').sort_values('Cantidad', ascending=False)
# Crear gráfico
ax = df_sorted.plot(kind='barh', x='Nombre del Producto', y='Cantidad', color= '#2842d5')
# Invertir eje y
ax.invert_yaxis()
# Esconder nombre del eje y
ax.set_ylabel('')
# Configurar eje x
ax.xaxis.set_major_locator(ticker.MaxNLocator(integer=True))
# Añadir título
plt.title('Top 10 Auriculares vendidos')
# Mostrar gráfico
plt.show()
```

```python
# Filtrar y ordenar datos
df_sorted = df[df['Categoria'] == 'Auricular'].nlargest(10, 'Ganancia').sort_values('Ganancia', ascending=False)
# Crear gráfico
ax = df_sorted.plot(kind='barh', x='Nombre del Producto', y='Ganancia', color= '#6B8E23')
# Invertir eje y
ax.invert_yaxis()
# Esconder nombre del eje y
ax.set_ylabel('')
# Configurar eje x
ax.xaxis.set_major_locator(ticker.MaxNLocator(integer=True))
# Añadir título
plt.title('Top 10 Auriculares más ganancia')
# Mostrar gráfico
plt.show()
```
